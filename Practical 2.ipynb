```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('emails.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.corr(numeric_only = True)
```

```python
df=df.drop('Email No.',axis=1)
```

```python
df
```

```python
X= df.drop('Prediction',axis=1)
```

```python
X
```

```python
y=df['Prediction']
```

```python
y
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X= scaler.fit_transform(X)
```

```python
X
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier()
```

```python
KNN_model= KNN.fit(X_train,y_train)
```

```python
KNN_predictions= KNN_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
KNN_accuracy= accuracy_score(y_test,KNN_predictions)
```

```python
KNN_accuracy
```

## SVC

```python
from sklearn.svm import SVC
```

```python
SVC = SVC()
```

```python
SVC_model = SVC.fit(X_train,y_train)
```

```python
SVC_predictions = SVC_model.predict(X_test)
```

```python
SVC_accuracy = accuracy_score(y_test,SVC_predictions)
```

```python
SVC_accuracy*100
```

## KNN Confusion Matrix

```python
KNN_cf= confusion_matrix(y_test,KNN_predictions)
KNN_cf
```

```python
sns.heatmap(KNN_cf, annot=True)
```

## SVC Confusion Matrix

```python
SVC_cf=confusion_matrix(y_test,SVC_predictions)
SVC_cf
```

```python
sns.heatmap(SVC_cf,annot=True )
```

## Precision v/s recall analysis

```python
KNN_precision= precision_score(y_test,KNN_predictions)
KNN_precision
```

```python
KNN_recall = recall_score(y_test,KNN_predictions)
KNN_recall
```

```python
SVC_precision= precision_score(y_test,SVC_predictions)
SVC_precision
```

```python
SVC_recall = recall_score(y_test,SVC_predictions)
SVC_recall
```
